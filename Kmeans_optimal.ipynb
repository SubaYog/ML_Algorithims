```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
cus.head()
```

```python
#2 Check the number of cells in each column with null values

cus.isnull().sum()
```

```python
#3 Create a scatter plot with Age as X and Spending Score as Y

plt.figure(1 , figsize = (15 , 7))
plt.title('Scatter plot of Age v/s Spending Score', fontsize = 20)
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.scatter( x = 'Age', y = 'Spending Score (1-100)', data = cus)
plt.show()
```

```python
#4.Find out the best number for clusters between 1 and 10 (inclusive) using the elbowmethod.
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
```

```python
lf = preprocessing.LabelEncoder()
cus['Gender'] = lf.fit_transform(cus['Gender'])
```

```python
cus.head()
```

```python
kmeans = KMeans(n_clusters = 4)
kmeans
```

```python
k_model = kmeans.fit(cus)
```

```python
k_model.n_clusters
```

```python
centers = k_model.cluster_centers_ #geting centres
centers
```

```python
k_model.labels_
```

```python
k_means = KMeans(n_clusters=2).fit(cus)
```

```python
cluster_ = k_means.labels_
cluster_
```

```python
plt.scatter(cus.iloc[:,1],cus.iloc[:,2],c=cluster_,s=50,cmap="viridis");
```

```python
T=[]
for k in range(1,11):
    kmeans = KMeans(n_clusters = k).fit(cus)
    T.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), T)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('T')
plt.show()
```
