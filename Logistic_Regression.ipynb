```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ar.drop(['Arrival Delay in Minutes'], axis = 1,inplace= True)
ar
```

```python
ar = pd.get_dummies(ar, columns=['satisfaction', 'Gender', 'Customer Type'
                                 , 'Type of Travel', 'Class'], drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler
stand= StandardScaler()
```

```python
sns.kdeplot(ar['Age'], shade=True)
sns.kdeplot(ar['Flight Distance'], shade=True)
sns.kdeplot(ar['Departure Delay in Minutes'], shade=True)
```

```python
for column in ['Age','Flight Distance','Departure Delay in Minutes']:
    ar[column] = stand.fit_transform(ar[column].values.reshape(-1,1))
```

```python
X =ar.drop(['satisfaction_satisfied'], axis = 1)
```

```python
y = ar['satisfaction_satisfied']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print("Accuracy: %s%%" % (100*accuracy_score(y_test, y_pred)))
print(confusion_matrix(y_test, y_pred))
```

```python
print("Classification Report for Logistic Regression")
print(classification_report(y_test, y_pred))
```
